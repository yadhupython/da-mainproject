### Introduction to Export to Asian Countries
The 'Exports to Asian Countries' dataset provides a comprehensive view of trade exports from an unspecified country (presumably India given the source) to various Asian countries. This dataset covers the periodical data of exports in terms of value in both USD and INR and also in terms of quantity. Each record provides the date, target country, type of product (both in terms of HS code and the commodity name), the value of the commodity, and the quantity of the commodity being exported. 

### Data Source
"The data has been sourced from the 'Trade Statistics' section of the commerce department's official website. Data is presumably collected through customs and trade declarations from businesses involved in the export trade, processed by the respective government department, and then made publicly available on the mentioned portal.(https://indiadataportal.com/p/export-trade-statistics/r/mci-tradestat_export_lfy-cn-mn-asi)"

Data cleaning!
1.Fix Data Types

```python
numeric_cols = [
    "value_qt", "value_rs", "value_dl",
    "country_code", "region_code", "sub_region_code", "hs_code"
]

df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors="coerce")
df["date"] = pd.to_datetime(df["date"], format="mixed", dayfirst=True)
df.info()
```

2.Clean Text Columns

```python
text_cols = ["country_name", "alpha_3_code", "region", "sub_region", "commodity", "unit"]

for col in text_cols:
    df[col] = df[col].astype(str).str.strip().str.title()
```

3.Convert “Thd” (thousand) to actual quantity

```python
df.loc[df["unit"] == "Thd", "value_qt"] *= 1000
df["unit"] = df["unit"].replace({"Thd": "Kgs"})
df.info()
```

4.Handle Missing or Zero Values

```python
df[["value_rs", "value_dl"]] = df[["value_rs", "value_dl"]].replace(0, pd.NA)
df["unit"] = df["unit"].replace('"', "Unknown")

df.info()
```

```python
df["value_rs"] = pd.to_numeric(df["value_rs"], errors="coerce")
df["value_dl"] = pd.to_numeric(df["value_dl"], errors="coerce")
```

5.Detect Outliers

```python
def detect_outliers(group):
    mean = group["value_qt"].mean()
    std = group["value_qt"].std()
    return (group["value_qt"] - mean).abs() > 3 * std

df["is_outlier"] = df.groupby("hs_code").apply(detect_outliers).reset_index(level=0, drop=True)
outliers = df[df["is_outlier"]]
```

Flaggeed Outliers

```python
def detect_outliers(group):
    mean = group["value_qt"].mean()
    std = group["value_qt"].std()
    return (group["value_qt"] - mean).abs() > 3 * std

df["is_outlier"] = (
    df.groupby("hs_code")
      .apply(detect_outliers)
      .reset_index(level=0, drop=True)
)
```

Inspect Outliers

```python
outliers = df[df["is_outlier"]]
outliers.head(20)
```

Remove Only Incorrect outliers

```python
df_clean = df[df["is_outlier"] == False]
```

5.Remove Duplicates

```python
df = df.drop_duplicates()
```

Validate HS Codes

```python
df = df[df["hs_code"].astype(str).str.len().between(6, 10)]
```

Clean Commodity Descriptions

```python
df["commodity"] = (
    df["commodity"]
    .str.replace(r"\s+", " ", regex=True)
    .str.replace(r"[^a-zA-Z0-9 ,\-()]", "", regex=True)
)
```

### Importing Required Libraries
Neccessary Python libraries such as Pandas,Plotly,Matplotli,and Seaborn are imported for numerical operations and data visualizations.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

Country vs Total Value (value_rs)

```python
# Calculate total export value per country
country_value = (
    df.groupby("country_name")["value_rs"]
      .sum()
      .sort_values(ascending=False)
)

# Keep only Top 10 countries
top10 = country_value.head(10)

plt.figure(figsize=(16,7))
sns.set_style("whitegrid")

sns.barplot(
    x=top10.index,
    y=top10.values,
    palette="viridis"
)

plt.title("Top 10 Countries by Total Trade Value (RS)", fontsize=20, weight="bold")
plt.xlabel("Country", fontsize=14)
plt.ylabel("Total Value (RS)", fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.tight_layout()
plt.show()
```

Insight:highest‑value trade relationships are not evenly distributed — instead, a few countries dominate the overall rupee value.

Percentage of Top 10 Export Value by Country

```python
# Calculate total export value per country
country_value = (
    df.groupby("country_name")["value_rs"]
      .sum()
      .sort_values(ascending=False)
)

# Keep only Top 10 countries
top10 = country_value.head(10)

plt.figure(figsize=(10,10))
plt.pie(
    top10,
    labels=top10.index,
    autopct="%1.1f%%",
    startangle=140,
    pctdistance=0.85
)

# Donut effect
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title("Top 10 Countries by Percentage of Export Value", fontsize=18, weight="bold")
plt.tight_layout()
plt.show()
```

Insight:The Top 10 countries contribute the majority of the total export value, showing a strong concentration of trade in a small number of markets. This indicates that export performance is heavily dependent on these key countries, while the remaining countries contribute only marginally.

Bottom of  10 Export Value by Country

```python
# Calculate total export value per country
country_value = (
    df.groupby("country_name")["value_rs"]
      .sum()
      .sort_values(ascending=False)
)

# Keep only LAST 10 countries
last10 = country_value.tail(10)

plt.figure(figsize=(10,10))
plt.pie(
    last10,
    labels=last10.index,
    autopct="%1.1f%%",
    startangle=140,
    pctdistance=0.85
)

# Donut effect
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title("Bottom 10 Countries by Percentage of Export Value", fontsize=18, weight="bold")
plt.tight_layout()
plt.show()
```

The bottom 10 countries contribute only a very small share of the total export value. This indicates weak trade activity or limited product demand in these markets. These countries may represent low‑priority markets or potential areas for future development if strategic expansion is planned.

Subregion Percentage Donut Chart

```python
subregion_value = (
    df.groupby("sub_region")["value_rs"]
      .sum()
)

subregion_percent = (subregion_value / subregion_value.sum()) * 100
```

```python
plt.figure(figsize=(10,10))
plt.pie(
    subregion_percent,
    labels=subregion_percent.index,
    autopct="%1.1f%%",
    startangle=140,
    pctdistance=0.85
)

centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title("Percentage of Export Value by Subregion", fontsize=18, weight="bold")
plt.tight_layout()
plt.show()
```

Insight:exports are heavily concentrated in a few subregions, while several other subregions contribute only a small percentage. This indicates that your export network is not evenly distributed, and your business relies more on specific subregional markets

Year‑wise Export Trend

```python
df["date"] = pd.to_datetime(df["date"])
df["year"] = df["date"].dt.year
```

```python
year_counts = (
    df.groupby("year")
      .size()
      .reset_index(name="export_count")
      .sort_values("year")
)

plt.figure(figsize=(12,6))
sns.lineplot(
    data=year_counts,
    x="year",
    y="export_count",
    marker="o",
    linewidth=3,
    color="royalblue"
)

plt.title("Year‑wise Export Activity (Transaction Count)", fontsize=18, weight="bold")
plt.xlabel("Year", fontsize=14)
plt.ylabel("Number of Export Transactions", fontsize=14)
plt.grid(True, linestyle="--", alpha=0.4)
plt.tight_layout()
plt.show()
```

The chart displays export transaction counts for 2024 and 2025, with a noticeable increase from roughly 375,000 to 445,000 transactions. The line rises steadily, indicating a positive shift in export activity.

Insight:export operations experienced a significant year‑over‑year increase, suggesting stronger market demand, improved operational efficiency, or expansion into new destinations. This upward movement signals healthy growth momentum in your export pipeline.

Top 10 Exporting Countries

```python
# Top 10 exporting countries by transaction count
top_countries = (
    df['country_name']
    .value_counts()
    .head(10)
    .reset_index()
)

top_countries.columns = ['country_name', 'export_count']

plt.figure(figsize=(12,6))
sns.barplot(
    data=top_countries,
    x='export_count',
    y='country_name',
    palette='viridis'
)

plt.title("Top 10 Exporting Countries", fontsize=18, weight='bold')
plt.xlabel("Number of Export Transactions", fontsize=14)
plt.ylabel("Country", fontsize=14)
plt.tight_layout()
plt.show()
```

Insights:The Top 10 Exporting Countries chart reveals that export activity is heavily concentrated among a small group of countries, indicating strong and established trade relationships with these markets.

Country‑Wise Export Growth Rate (Top 10 Countries Only)

```python
# 1. Identify top 10 exporting countries
top_10_countries = (
    df['country_name']          # <-- replace with your actual country column
    .value_counts()
    .head(10)
    .index
)

# 2. Filter dataset for only these top 10 countries
df_top10 = df[df['country_name'].isin(top_10_countries)]

# 3. Group by country and year to calculate export counts
growth_data = (
    df_top10.groupby(['country_name', 'year'])
            .size()
            .reset_index(name='export_count')
)

# 4. Plot growth rate
plt.figure(figsize=(14,7))
sns.lineplot(
    data=growth_data,
    x='year',
    y='export_count',
    hue='country_name',
    marker='o'
)

plt.title("Country‑Wise Export Growth Rate (Top 10 Countries)", fontsize=18, weight='bold')
plt.xlabel("Year", fontsize=14)
plt.ylabel("Number of Export Transactions", fontsize=14)
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

Insights:The growth rate visualization for the top 10 exporting countries shows clear differences in how export activity evolves across major markets. Some countries demonstrate consistent year‑over‑year growth, indicating strengthening trade relationships, while others show fluctuating patterns that may require strategic attention. 

Top 10 Fastest Growing Export Countries

```python
# 1. Calculate yearly export count per country
yearly_counts = (
    df.groupby(['country_name', 'year'])
      .size()
      .reset_index(name='export_count')
)

# 2. Calculate year-over-year growth for each country
yearly_counts['growth_rate'] = yearly_counts.groupby('country_name')['export_count'].pct_change()

# 3. Compute average growth rate per country
avg_growth = (
    yearly_counts.groupby('country_name')['growth_rate']
                 .mean()
                 .reset_index()
                 .dropna()
)

# 4. Select top 10 fastest growing countries
top_fastest = avg_growth.sort_values(by='growth_rate', ascending=False).head(10)

# 5. Plot the fastest growing countries
plt.figure(figsize=(12,6))
sns.barplot(
    data=top_fastest,
    x='growth_rate',
    y='country_name',
    palette='crest'
)

plt.title("Top 10 Fastest Growing Export Countries", fontsize=18, weight='bold')
plt.xlabel("Average Growth Rate (YoY)", fontsize=14)
plt.ylabel("Country", fontsize=14)
plt.tight_layout()
plt.show()
```

Insight:The fastest‑growing export countries show strong upward momentum, indicating rapidly expanding trade relationships and increasing demand from these markets. These countries outperform others in year‑over‑year growth, making them high‑potential targets for future investment and strategic focus. 

Underperforming Countries Heatmap (Bottom 20 Countries)

```python
# 1. Identify bottom 20 underperforming countries
bottom_20 = (
    df['country_name']     # <-- replace with your actual country column
    .value_counts()
    .sort_values(ascending=True)
    .head(20)
    .index
)

# 2. Filter dataset for only these countries
df_bottom20 = df[df['country_name'].isin(bottom_20)]

# 3. Create pivot table for heatmap
heatmap_data = df_bottom20.pivot_table(
    index='country_name',
    columns='year',
    aggfunc='size',
    fill_value=0
)

# 4. Plot heatmap
plt.figure(figsize=(14,10))
sns.heatmap(
    heatmap_data,
    cmap='Reds',
    linewidths=0.5
)

plt.title("Underperforming Countries Heatmap (Bottom 20 Exporters)", fontsize=18, weight='bold')
plt.xlabel("Year", fontsize=14)
plt.ylabel("Country", fontsize=14)
plt.tight_layout()
plt.show()
```

Insight:The heatmap reveals that the bottom 20 countries show consistently low export activity across most years, indicating weak or underdeveloped trade relationships. Some countries show occasional spikes, suggesting sporadic opportunities, while others remain persistently inactive. 

Sub‑Region Wise Export Comparison

```python
# Sub-region wise export count comparison

subregion_count = (
    df['sub_region']        # <-- your subregion column
    .value_counts()
    .reset_index()
)

subregion_count.columns = ['sub_region', 'export_count']

plt.figure(figsize=(12,6))
sns.barplot(
    data=subregion_count,
    x='export_count',
    y='sub_region',
    palette='coolwarm'
)

plt.title("Sub‑Region Wise Export Count Comparison", fontsize=18, weight='bold')
plt.xlabel("Total Export Count", fontsize=14)
plt.ylabel("Sub‑Region", fontsize=14)
plt.tight_layout()
plt.show()
```
